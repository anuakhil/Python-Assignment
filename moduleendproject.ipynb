```python
#Preprocessing

import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("C:\\Users\\USER\\OneDrive\\Desktop\\Python\\project.csv")

# Check the height column for issues
df['Height'] = df['Height'].apply(lambda x: np.random.randint(150, 181) if pd.isna(x) or not isinstance(x, (int, float)) else x)

# Save the cleaned dataset
df.to_csv("cleaned_project.csv", index=False)
```

```python
#Analysis Tasks

#Task 1

# Group by team and count employees
team_distribution = df['Team'].value_counts()

# Calculate percentage
team_percentage = (team_distribution / len(df)) * 100

# Combine the results into a DataFrame
team_analysis = pd.DataFrame({
    'Employee Count': team_distribution,
    'Percentage': team_percentage
})
team_analysis
```

```python
#Task 2

# Group by position and count employees
position_distribution = df['Position'].value_counts()

# Display results
position_distribution
```

```python
#Task 3

# Assuming age groups are defined in a column 'age_group' or by age ranges
age_group_distribution = df['Age'].value_counts()

# Get the most common age group
predominant_age_group = age_group_distribution.idxmax()
predominant_age_group
```

```python
#Task 4

# Group by team and sum salaries
team_salary_expenditure = df.groupby('Team')['Salary'].sum()

# Group by position and sum salaries
position_salary_expenditure = df.groupby('Position')['Salary'].sum()

# Find the team and position with the highest expenditure
highest_team_salary = team_salary_expenditure.idxmax(), team_salary_expenditure.max()
highest_position_salary = position_salary_expenditure.idxmax(), position_salary_expenditure.max()

highest_team_salary, highest_position_salary
```

```python
#Task 5

import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation
correlation = df['Age'].corr(df['Salary'])

# Create a scatter plot to visualize the correlation
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title(f"Correlation between Age and Salary: {correlation:.2f}")
plt.show()
```

```python
#Graphical Representation

team_analysis['Employee Count'].plot(kind='bar', title='Employee Distribution Across Teams')
plt.xlabel('Team')
plt.ylabel('Employee Count')
plt.show()

team_analysis['Percentage'].plot(kind='bar', title='Percentage Distribution Across Teams')
plt.xlabel('Team')
plt.ylabel('Percentage')
plt.show()
```

```python
# Data Story

#Segregation of employees based on positions

position_distribution.plot(kind='bar', title='Employee Distribution Across Positions')
plt.xlabel('Position')
plt.ylabel('Employee Count')
plt.show()
```

```python
#Predominant age group

age_group_distribution.plot(kind='pie', autopct='%1.1f%%', title='Predominant Age Group')
plt.ylabel('')
plt.show()
```

```python
#Highest salary expenditure by team and position

team_salary_expenditure.plot(kind='bar', title='Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.show()

position_salary_expenditure.plot(kind='bar', title='Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary Expenditure')
plt.show()
```
